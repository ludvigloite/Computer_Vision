## Task 2e)


As is apparent in both the validation loss and the train accuracy, our model suffers greatly from performance spikes once every sixth epoch.

We believe these spikes are caused by bad images in one of the 32 batches of our training data set, which makes sense considering the spike in training loss once every epoch. Since validation is performed on an entirely different data set only once every fifth training step, this problem does not become apparent until we run a validation step right after training our model on this bad batch. With 32 batches in our training data, and running validation once every fifth step, this accounts for the performance spike in validation loss and accuracy once every 192th step.

By introducing data set shuffling, we can avoid this problem to some extent. We will of course still suffer from the same bad data, but spread over different batches every epoch such that the model is not consistently updated with bad weights at regular intervals. Additionally, since shuffling the data set allows for more consistent validation loss results, we can confidently stop training out model much earlier to avoid overfitting. (Siste setning ok?)
